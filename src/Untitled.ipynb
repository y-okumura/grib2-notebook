```python
from urllib.request import urlretrieve
urlretrieve(url, 'grib2.dat')
```

```python
import pygrib

gribs = pygrib.open('grib2.dat')

for i in range(1, gribs.messages):
  print(gribs[i])
```

```python
totalCloudCover = gribs.select(name='Total Cloud Cover',typeOfLevel='isobaricInhPa',)
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature

from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt

for row in totalCloudCover:
    print(f'{row.level} hPa')

    # 幅と高さから出力ファイルの大きさを設定。DPIで画像の大きさを調整する
    plt.figure(figsize=(30, 20), dpi=50)

    # 取得したデータの中心を中心にした正射方位図法で出力し、
    ax = plt.axes(projection=ccrs.Orthographic(135, 35))
    ax.coastlines(resolution='50m')
    ax.gridlines(linestyle='-', color='white')
    
    
    (data, lats, lons) = row.data(lat1=30, lat2=40, lon1=130, lon2=140)
    plt.contourf(
        lons, lats, data,
        vmin=0,
        vmax=100,
        transform = ccrs.PlateCarree(),
        cmap='PuBu_r',
        alpha=0.5,
        extend='neither',
        antialiased=True
    )
    plt.show()
```
